```python
import qiskit as q
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt
```

# Porte de Hadamard Multiple

```python
n=2
N=2**n

simulator = q.Aer.get_backend('qasm_simulator')
circuit = q.QuantumCircuit(N, N)
```

```python
# Création du circuit
L = []

for i in range(N):
    circuit.h(i)

    L.append(i) # Permet d'obtenir une mesure sur chaque qubit

circuit.measure(L, L) # Mesures
circuit.draw(output='mpl')
```

```python
# Exécution de l'algorithme 1000 fois
job = q.execute(circuit, simulator, shots=1000)

# Résultats des mesures
result = job.result()

counts = result.get_counts(circuit)
print(counts)
```
